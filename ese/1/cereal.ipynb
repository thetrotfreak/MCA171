## Import the dataset

```python
import pandas as pd

ds = pd.read_csv("./UScereal - UScereal.csv")

print(ds.info())
```

## Summary Statistics of Dataset

```python
print(ds.describe())

print(ds.isnull().sum())  # No Nulls are present in Dataset
```

## Average Protein Value

```python
ds.groupby("mfr").protein.mean()
```

## High Sugar

```python
ds.groupby("mfr").sugars.mean().sort_values(ascending=False)
```

## Plots

```python
ds.sample(10).plot.bar(x="vitamins")
```

```python
ds[["mfr", "fibre"]].head().plot.barh(x="mfr", y="fibre")
```

```python
ds.plot.scatter(x="sugars", y="sodium")
```
