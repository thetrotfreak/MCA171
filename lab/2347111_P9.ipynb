```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_csv("../dataset/goodreads_books.csv")

df.info()
print("\n\n")
df.describe()
```

## Histogram

```python
df.hist(column="num_pages")

plt.title("Histogram for Number of Pages")
plt.xlabel("Pages")
plt.ylabel("Frequency")

plt.show()
```

## Scatterplot

```python
df.plot.scatter(x="ratings_count", y="text_reviews_count")
```

## Line Plot

```python
df.plot.line(x="ratings_count", y="text_reviews_count")
```

## Bargraph

```python
sample = df.sample(10)
sample.plot(kind="bar")
```

## Pie Chart

```python
plt.pie(sample["num_pages"])
plt.show()
```

## Area Plot

```python
df.plot.area()
```

## Box Plot

```python
plt.figure(figsize=(10, 7))
plt.boxplot(df["average_rating"])
plt.show()
```

## Pair Plot

```python
sb.pairplot(df.head(10))
```

## Stem Plot

```python
plt.stem(sample["average_rating"], sample["num_pages"])
plt.show()
```

## Fill Graph

```python
plt.fill(sample["average_rating"], sample["num_pages"])
plt.show()
```
